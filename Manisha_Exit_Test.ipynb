# Preprocessing for dataset

Finding the missing values in each column of dataset

```python
data.isnull().sum()
```

Handling the missing values

```python
for i in ['gender']:
    data[i]=data[i].fillna(value='Unknown')
```

```python
for i in ['weight']:
    data[i]=data[i].fillna(value='Unknown')
```

```python
for i in ['ball_controlling_skills']:
    data[i]=data[i].fillna(data[i].median())
```

```python
for i in ['jumping_skills']:
    data[i]=data[i].fillna(data[i].median())
for i in ['penalties_conversion_rate']:
    data[i]=data[i].fillna(data[i].median())
for i in ['mental_strength']:
    data[i]=data[i].fillna(data[i].median())
for i in ['shot_accuracy']:
    data[i]=data[i].fillna(data[i].median())
for i in ['behaviour_rating']:
    data[i]=data[i].fillna(data[i].mean())
for i in ['matches_played']:
    data[i]=data[i].fillna(data[i].median())
for i in ['fitness_rating']:
    data[i]=data[i].fillna(data[i].mean()) 
for i in ['years_of_experience']:
    data[i]=data[i].fillna(data[i].mean()) 
```

```python
data.isnull().sum()
```

```python
for i in ['strong_foot']:
    data[i]=data[i].fillna(value='Unknown')
for i in ['coaching']:
    data[i]=data[i].fillna(value='Unknown')
```

```python
data.isnull().sum()
```

# Modelling and Fine Tunning

```python
data.columns
```

```python
x=data.drop(['player_id', 'name','coaching','strong_foot','weight','dedication_level','height'],axis=1)
y=data['selection']
```

```python
x.head()
```

```python
x.isna().sum()
```

```python
x.dtypes
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
from sklearn.svm import SVC
svmclf=SVC(kernel = 'linear')
svmclf.fit(x_train,y_train)
```

```python
y_pred=svmclf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy is ',accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
