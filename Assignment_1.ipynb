```python
# Pie Chart

y=data['Classification'].value_counts()
mylabels=data['Classification'].unique().tolist()
```

```python
plt.pie(y,labels=mylabels,radius=1)
plt.show()
print(y)
```

```python
#Correlation 

corr_matrix=data.corr()
sns.heatmap(corr_matrix,annot=True,cmap='PiYG')
```

```python
#Scatter based on classification

colors=['Red','Green','Blue']
Classification=['Iris-setosa','Iris-versicolor','Iris-virginica']   
```

```python
for i in range(3):
    x=data[data['Classification']==Classification[i]]
    plt.scatter(x['SL'],x['SW'],color=colors[i],label=Classification[i])
    plt.xlabel("SL")
    plt.ylabel("SW")
    plt.legend()       
```
